```python
# Data Manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Plotly
import plotly.express as px
import plotly.graph_objects as go

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Nassau Candy Distributor.csv")

print("Shape:", df.shape)

df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    dayfirst=True,
    errors='coerce'
)

df['Ship Date'] = pd.to_datetime(
    df['Ship Date'],
    dayfirst=True,
    errors='coerce'
)
```

```python
df.isnull().sum()
```

```python
df['Lead Time'] = (
    df['Ship Date'] -
    df['Order Date']
).dt.days
```

```python
df[['Order Date','Ship Date','Lead Time']].head()
```

```python
df = df[df['Lead Time'] >= 0]

print(df.shape)
```

```python
factory_map = {
    "Wonka Bar - Nutty Crunch Surprise":"Lot's O' Nuts",
    "Wonka Bar - Fudge Mallows":"Lot's O' Nuts",
    "Wonka Bar -Scrumdiddlyumptious":"Lot's O' Nuts",
    "Wonka Bar - Milk Chocolate":"Wicked Choccy's",
    "Wonka Bar - Triple Dazzle Caramel":"Wicked Choccy's",
    "Laffy Taffy":"Sugar Shack",
    "SweeTARTS":"Sugar Shack",
    "Nerds":"Sugar Shack",
    "Fun Dip":"Sugar Shack",
    "Fizzy Lifting Drinks":"Sugar Shack",
    "Everlasting Gobstopper":"Secret Factory",
    "Hair Toffee":"The Other Factory",
    "Lickable Wallpaper":"Secret Factory",
    "Wonka Gum":"Secret Factory",
    "Kazookles":"The Other Factory"
}
```

```python
df['Factory'] = df['Product Name'].map(factory_map)
```

```python
df['Factory'].value_counts()
```

```python
df['Route_State'] = (
    df['Factory']
    + " → "
    + df['State/Province']
)
```

```python
df['Route_Region'] = (
    df['Factory']
    + " → "
    + df['Region']
)
```

```python
delay_threshold = 5

df['Delayed'] = (
    df['Lead Time'] > delay_threshold
)
```

```python
total_orders = df['Order ID'].nunique()
total_sales = df['Sales'].sum()
total_profit = df['Gross Profit'].sum()
avg_lead_time = df['Lead Time'].mean()

delay_rate = (
    df['Delayed'].mean()
)*100

print("Total Orders:", total_orders)
print("Total Sales:", round(total_sales,2))
print("Total Profit:", round(total_profit,2))
print("Average Lead Time:", round(avg_lead_time,2))
print("Delay Rate:", round(delay_rate,2), "%")
```

```python
route_summary = df.groupby(
    'Route_State'
).agg(
    Shipments=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Median_Lead_Time=('Lead Time','median'),
    Std_Lead_Time=('Lead Time','std'),
    Delay_Frequency=('Delayed','mean'),
    Sales=('Sales','sum'),
    Profit=('Gross Profit','sum')
).reset_index()
```

```python
min_lt = route_summary['Avg_Lead_Time'].min()
max_lt = route_summary['Avg_Lead_Time'].max()
```

```python
route_summary['Efficiency Score'] = (
    100 *
    (
        1 -
        (
            (route_summary['Avg_Lead_Time'] - min_lt)
            /
            (max_lt - min_lt)
        )
    )
)
```

```python
top10 = route_summary.sort_values(
    'Avg_Lead_Time'
).head(10)

top10
```

```python
bottom10 = route_summary.sort_values(
    'Avg_Lead_Time',
    ascending=False
).head(10)

bottom10
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='Avg_Lead_Time',
    y='Route_State'
)

plt.title('Top 10 Most Efficient Routes')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=bottom10,
    x='Avg_Lead_Time',
    y='Route_State'
)

plt.title('Bottom 10 Least Efficient Routes')
plt.show()
```

```python
factory_summary = df.groupby(
    'Factory'
).agg(
    Orders=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Delay_Rate=('Delayed','mean'),
    Sales=('Sales','sum'),
    Profit=('Gross Profit','sum')
).reset_index()

factory_summary
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=factory_summary,
    x='Factory',
    y='Avg_Lead_Time'
)

plt.xticks(rotation=45)
plt.show()
```

```python
region_summary = df.groupby(
    'Region'
).agg(
    Orders=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Delay_Rate=('Delayed','mean'),
    Sales=('Sales','sum')
).reset_index()

region_summary
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=region_summary,
    x='Region',
    y='Avg_Lead_Time'
)

plt.show()
```

```python
state_summary = df.groupby(
    'State/Province'
).agg(
    Orders=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Delay_Rate=('Delayed','mean')
).reset_index()

state_summary.head()
```

```python
state_summary.sort_values(
    'Avg_Lead_Time',
    ascending=False
).head(15)
```

```python
shipmode_summary = df.groupby(
    'Ship Mode'
).agg(
    Orders=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Sales=('Sales','sum'),
    Profit=('Gross Profit','sum')
).reset_index()

shipmode_summary
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=shipmode_summary,
    x='Ship Mode',
    y='Avg_Lead_Time'
)

plt.xticks(rotation=30)
plt.show()
```

```python
df['Order_Month'] = (
    df['Order Date']
    .dt.to_period('M')
    .astype(str)
)
```

```python
monthly = df.groupby(
    'Order_Month'
).agg(
    Orders=('Order ID','count'),
    Sales=('Sales','sum')
).reset_index()

monthly.head()
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    monthly['Order_Month'],
    monthly['Orders']
)

plt.xticks(rotation=90)
plt.title('Monthly Orders Trend')
plt.show()
```

```python
volume_threshold = state_summary[
    'Orders'
].quantile(0.75)

lead_threshold = state_summary[
    'Avg_Lead_Time'
].mean()
```

```python
bottlenecks = state_summary[
    (state_summary['Orders'] > volume_threshold)
    &
    (state_summary['Avg_Lead_Time'] > lead_threshold)
]

bottlenecks
```

```python
route_summary.to_csv(
    'Route_Summary.csv',
    index=False
)

factory_summary.to_csv(
    'Factory_Summary.csv',
    index=False
)

state_summary.to_csv(
    'State_Summary.csv',
    index=False
)

region_summary.to_csv(
    'Region_Summary.csv',
    index=False
)
```
